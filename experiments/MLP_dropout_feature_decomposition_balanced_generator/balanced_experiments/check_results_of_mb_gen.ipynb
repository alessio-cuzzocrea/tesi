```python
import pandas as pd

data_dir = "data/"
report_data_folder = "data_for_report/"
```

```python
cv_results_mb_gen_maxnorm = pd.read_csv(data_dir + "cv_results_scaler_adam_with_dropout_mb_gen.csv")
df = cv_results_mb_gen_maxnorm.sort_values("mean_test_AU_PRC", ascending=False)[["mean_test_AVG_PREC","std_test_AVG_PREC","mean_train_AVG_PREC","mean_test_AU_ROC","mean_test_AU_PRC" ]]
df = df.round(3)
df = df.rename(lambda x: x.replace("_", ""), axis=1)
df.to_csv(report_data_folder + "cv_results_scaler_adam_with_dropout_mb_gen.csv", index=False)
df
```
