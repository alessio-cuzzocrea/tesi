```python
import pandas as pd

data_dir = "data/"
data_report_folder = "data_for_report/"
```

```python
cv_results_dropout_train = pd.read_csv(data_dir + "cv_results_scaler_adam_dropout_top_ten_train.csv")
cv_results_dropout_test = pd.read_csv(data_dir + "cv_results_scaler_adam_dropout_top_ten_test.csv")
```

```python
cv_results_dropout_test.sort_values("mean_test_AVG_PREC", ascending=False)[["param_model__architecture", "param_model__dropout_rate", "param_model__epochs","mean_test_AU_PRC","mean_test_AVG_PREC","mean_test_AU_ROC","mean_train_AVG_PREC", "std_test_AU_PRC"]].head()
```

# Cleaning data for report

```python
df = cv_results_dropout_train.sort_values("mean_test_AVG_PREC", ascending=False)[["param_model__architecture","param_model__epochs","param_model__dropout_rate","mean_test_AVG_PREC","mean_train_AVG_PREC", "std_test_AVG_PREC", "mean_fit_time"]]
df = df.round(3)
df = df.rename(lambda x: x.replace("_", ""), axis=1)
df = df.rename(lambda x: x.replace("parammodelarchitecture", "architecture"), axis=1)
df = df.rename(lambda x: x.replace("parammodeldropoutrate", "dropoutrate"), axis=1)
df.architecture = df.architecture.apply(lambda x: x.replace("(", ""))
df.architecture = df.architecture.apply(lambda x: x.replace(")", ""))
df.architecture = df.architecture.apply(lambda x: x.replace(",", "") if x[-1] == "," else x)
df.architecture = df.architecture.apply(lambda x: x.replace(", ", "-"))
df = df[df.parammodelepochs == 150]
df.to_csv(data_report_folder + "adam_dropout_scaling_top_ten_train.csv", index=False)
df.head(5)
```

```python
non_overlapping_architectures = list(set(cv_results_dropout_test.param_model__architecture) - set (cv_results_dropout_train.param_model__architecture))
```

```python
cv_results_dropout_test = cv_results_dropout_test[cv_results_dropout_test.param_model__architecture.isin(non_overlapping_architectures)]
df = cv_results_dropout_test.sort_values("mean_test_AVG_PREC", ascending=False)[["param_model__architecture","param_model__epochs","param_model__dropout_rate","mean_test_AVG_PREC","mean_train_AVG_PREC", "std_test_AVG_PREC", "mean_fit_time"]]
df = df.round(3)
df = df.rename(lambda x: x.replace("_", ""), axis=1)
df = df.rename(lambda x: x.replace("parammodelarchitecture", "architecture"), axis=1)
df = df.rename(lambda x: x.replace("parammodeldropoutrate", "dropoutrate"), axis=1)
df.architecture = df.architecture.apply(lambda x: x.replace("(", ""))
df.architecture = df.architecture.apply(lambda x: x.replace(")", ""))
df.architecture = df.architecture.apply(lambda x: x.replace(",", "") if x[-1] == "," else x)
df.architecture = df.architecture.apply(lambda x: x.replace(", ", "-"))
df = df[df.parammodelepochs == 150]
df.to_csv(data_report_folder + "adam_dropout_scaling_top_ten_test.csv", index=False)
df.head(5)
```
