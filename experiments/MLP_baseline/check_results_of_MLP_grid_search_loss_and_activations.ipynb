```python
import pandas as pd
```

```python
cv_results = pd.read_csv("cv_results.csv")
```

```python
cv_results.sort_values("mean_test_AU_PRC", inplace=True, ascending=False)
cv_results[["mean_test_AU_PRC", "mean_test_AVG_PREC","mean_test_AU_ROC", "param_loss", "param_activation","mean_fit_time"]]
```

```python
sliced_results = cv_results[["mean_test_AU_PRC", "mean_test_AVG_PREC","mean_test_AU_ROC", "params", "mean_fit_time"]]
series = sliced_results["params"].apply(lambda x: x.replace("<function hingesig at 0x7f7397f42bf8>", "'hinge'"))
series = series.apply(lambda x: eval(x)).apply(pd.Series)
sliced_results = pd.concat([series, sliced_results], axis=1)
sliced_results = sliced_results.drop("params", axis=1)
sliced_results = sliced_results.rename(lambda x: x.replace("_", ""), axis=1)
sliced_results = sliced_results.round(decimals=3)
sliced_results = sliced_results.applymap(lambda x: x.replace("_", " ") if isinstance(x, str) else x)
sliced_results.to_csv("cv_results_for_report.csv", index=False)
```
