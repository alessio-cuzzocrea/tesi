```python
import numpy as np
np.random.seed(2024)
from scipy.stats import binom
from bioinformatics_helpers.utils import get_mendelian_dataset
import matplotlib.pyplot as plt
```

```python
train_X, _,_,_ = get_mendelian_dataset()
batch_size = 5000
```

```python
p = 356 / len(train_X)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,8))
n = np.arange(0, 10000)


prob_gte_1 = 1 - binom.pmf(0, batch_size, p)

ax.vlines(5000,-1,np.round(prob_gte_1, 2) ,linestyle="-", color="firebrick", linewidth=1)
ax.hlines(np.round(prob_gte_1, 2), -200, batch_size,  linestyle="-", color="firebrick", linewidth=1)
ax.plot(n, (1 - binom.pmf(0, n, p)), ms=8, label='binom pmf')

ticks = ax.get_xticks()
ax.set_xticks(np.append(ticks, batch_size))
ticks = ax.get_yticks()
ax.set_yticks(np.append(ticks, np.round(prob_gte_1, 2)))
ax.grid()
ax.set_ylim(-0.02, 1.02)
ax.set_xlim(-200, 10200)
ax.set_xlabel("Size of a minibatch")
ax.set_ylabel("Probability to have at leaste a positive")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,8))
n = np.arange(0, 10000)

prob_gte_1 = 1 - binom.pmf(0, batch_size, p)

ax.vlines(5000,-1,np.round(prob_gte_1, 2) ,linestyle="-", color="firebrick", linewidth=1)
ax.hlines(np.round(prob_gte_1, 2), -200, batch_size,  linestyle="-", color="firebrick", linewidth=1)
ax.plot(n, 1 - np.power((1 - p),n), ms=8, label='binom pmf')

ticks = ax.get_xticks()
ax.set_xticks(np.append(ticks, batch_size))
ticks = ax.get_yticks()
ax.set_yticks(np.append(ticks, np.round(prob_gte_1, 2)))
ax.grid()
ax.set_ylim(-0.02, 1.02)
ax.set_xlim(-200, 10200)
ax.set_xlabel("Size of a minibatch")
ax.set_ylabel("Probability to have at leaste a positive")
```
