```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
def plot_training_test_loss_history(loss_history, test_history, train_history, params):
    f, (ax1,ax2,ax3) = plt.subplots(3,sharex=False, sharey=False, gridspec_kw={"hspace":0.5}, figsize=(10,10))
    n_epochs = list(range(1, len(data["loss"]) + 1))
    ax1.grid()
    ax1.set_xlabel("Epochs")
    ax1.set_ylabel("Training loss")
    ax1.plot(n_epochs, loss_history)
    ax2.grid()
    ax2.set_ylabel("test AUPRC")
    ax2.set_xlabel("Epochs")
    ax2.plot(n_epochs, test_history)
    ax3.grid()
    ax3.set_ylabel("train AUPRC")
    ax3.set_xlabel("Epochs")
    ax3.plot(n_epochs, train_history)
    f.suptitle("psp={} npr={} np={}".format(params["positive_sample_perc"], params["np_ratio"], params["negative_perc"]))
    plt.show()
    filename = "history_plot_for_npr_{}_psp_{}_np_{}.png".format(
        str(params["np_ratio"]).replace(".", "-"),
        str(params["positive_sample_perc"]).replace(".", "-"),
        str(params["negative_perc"]).replace(".", "-")
        )
    images_folder = "images"
    f.savefig(os.path.join(images_folder, filename))
    
```

```python
cv_results_sample_negative = pd.read_csv("../data/cv_results_mb_gen.csv")
cv_results_sample_negative.sort_values("mean_test_AU_PRC", inplace=True, ascending=False)
sliced_results = cv_results_sample_negative[["params"]]
series = sliced_results["params"].apply(lambda x: eval(x)).apply(pd.Series)
sliced_results = pd.concat([series, sliced_results], axis=1)
sliced_results = sliced_results.drop("params", axis=1)
top_5 = sliced_results.head()
top_5
```

```python
params = top_5.iloc[0].to_dict()
base_data_folder = "300_epochs_data"
filename = "MLP_gen_npr_{}_psp_{}_np_{}.csv".format(
        params["np_ratio"],
        params["positive_sample_perc"],
        params["negative_perc"]
        )
data = pd.read_csv(os.path.join(base_data_folder, filename))
plot_training_test_loss_history(data["loss"], data["test_history"], data["train_history"], params)
```
